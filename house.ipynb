```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("data.csv")
```

```python
df
```

```python
import seaborn as sns
```

```python
sns.jointplot(x="sqft_living",y="price",data=df)
```

```python
sns.lmplot(x="sqft_living",y="price",data=df)
```

```python
y=df['price']
X=df[['bedrooms','bathrooms','sqft_living','sqft_lot','floors','condition','sqft_above','sqft_basement','yr_renovated']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.999)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.coef_)
```

```python
predictions=lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('y_test')
plt.ylabel('predictions')
```

```python
ans=lm.score(X_test,y_test)
print(ans*100)
```
